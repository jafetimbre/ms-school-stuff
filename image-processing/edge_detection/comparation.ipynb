# Compararea metodelor de detectie a marginilor

```python
import matplotlib.image as m_img
from matplotlib import pyplot as plt

imgs = {
    "lena_original" : m_img.imread("https://raw.githubusercontent.com/jafetimbre/ms-school-stuff/master/image-processing/edge_detection/res/lena.png"),
    "lena_sobel" : m_img.imread("https://raw.githubusercontent.com/jafetimbre/ms-school-stuff/master/image-processing/edge_detection/res/results/lena_sobel.png"),
    "lena_canny" : m_img.imread("https://raw.githubusercontent.com/jafetimbre/ms-school-stuff/master/image-processing/edge_detection/res/results/lena_canny.png"),
    "lena_aco" : m_img.imread("https://raw.githubusercontent.com/jafetimbre/ms-school-stuff/master/image-processing/edge_detection/res/results/lena_aco.png"),
    "tiger_original" : m_img.imread("https://raw.githubusercontent.com/jafetimbre/ms-school-stuff/master/image-processing/edge_detection/res/tiger.png"),
    "tiger_sobel" : m_img.imread("https://raw.githubusercontent.com/jafetimbre/ms-school-stuff/master/image-processing/edge_detection/res/results/tiger_sobel.png"),
    "tiger_canny" : m_img.imread("https://raw.githubusercontent.com/jafetimbre/ms-school-stuff/master/image-processing/edge_detection/res/results/tiger_canny.png"),
    "tiger_aco" : m_img.imread("https://raw.githubusercontent.com/jafetimbre/ms-school-stuff/master/image-processing/edge_detection/res/results/tiger_aco.png"),
}
```

```python
figure, axis = plt.subplots(2, 4, figsize=(30,15))

for i, el in enumerate(imgs.items()):
    name, img = el
    x = 0 if i < 4 else 1
    y = i if i < 4 else i - 4

    axis[x, y].imshow(img)
    axis[x, y].set_title(name)

figure.show()
```
